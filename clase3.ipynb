```python
df= pd.read_csv('datasets/cwurData.csv')
df.head(4)
```

```python
def create_category(x):
    if (x >= 1) & (x <= 100):
        return 'First Tier'
    elif (x > 100) & (x <= 200):
        return 'Second Tier'
    elif (x > 200) & (x <= 300):
        return 'Third Tier'
    return 'Other Top University'

df['Rank_Level'] = df['world_rank'].apply(lambda y:create_category(y))
df.head(5)
```

```python
myPivot = df.pivot_table(values = 'score',
                         index = 'country',
                         columns = 'Rank_Level'
                        )
```
